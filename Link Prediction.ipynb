A priori se cumple la hipótesis de Fernando, a una baja desviación del vector relación y una alta disperión de los nodos "destino" se consigue una alta predicción.A continuación, fijamos |w| = 200000, mode= "normal", ndim = 200 y vamos variando l(tamaño de ventana):

```python
e = experiment("asterix",7474,"neo4j","pytpytpyt","nom","normal","l",0.23)
e.run(1,20,1)
show(e.p)
```

Las mejores predicciones se consiguen con l = 1 , puede ser porque las relaciones que estamos tratando de predecir son a distancia 1, se deben hacer más pruebas. A continuación, fijamos |w| = 200000, mode= "normal", l = 6 y vamos variando ndim:

```python
e = experiment("asterix",7474,"neo4j","pytpytpyt","nom","normal","ndim",0.23)
e.run(1,61,10)
show(e.p)
```

Se observa como a partir de una cantidad determinada de dimensiones (50 aprox.) se consiguen las mejores predicciones y el comportamiento se mantiene para cualqueir número superior de dimensiones testeado. A continuación experimentos con mode="degree":

```python
e = experiment("asterix",7474,"neo4j","pytpytpyt","nom","degree","ns",0.23)
e.run([],1,6,100000)
show(e.p)
```

Los resultados son similares al modo "normal" pero las predicciones más óptimas se consiguen con un valor más pequeño de |(c,W)|.
